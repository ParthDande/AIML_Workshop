```python
#to work with missing values
#when we use the dropna() function we get on an averge 10% data loss
df.dropna()
```

```python
#we can replace the value the value in NaN
#fillna() function will replace the values in the null space in that particular column
# we have to work for each column
# we cannot use the fillna() function for entire dataframe at once.
df['marks'].fillna(60)
```

```python
#we can replace the values based on the type of the data
# if the values are discrete we can replace it with the median value
# if the values are continous then we replace it with average value.

#here the marks are continous
df.fillna(df['marks'].mean())
```

```python
#here  the values are discrete hence we replace the values here with median
df['age'].fillna(df['age'].median())
```

```python
df['class'].mode()#this will give the most frequent value
#then we will replace this value in the categorical data
```

```python
df['class'].fillna(df['class'].mode()[0])
```

```python
#we can also replace the previous value in the NaN
#this can be done by ffill() function- Forward Fill method.
#we also have bfill() function.#by this we can Fill the NaN value with the previous record.
df.ffill()
```
