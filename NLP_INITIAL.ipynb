```python
#task: to calcaulate the average of the words which appearing in the sentence
text = "They told that their ages are 21 25 and 23 respectively"
words = text.split()
ages = [int(word) for word in words if word.isdigit()]
average_age = sum(ages) / len(ages)
print(average_age)
#another approach
print(sum(ages)/len(ages))
#the algorithm will give the right answer but if there is a comma "," seperating the numbers then it will consider it as string
```

```python
from nltk.tokenize import word_tokenize,sent_tokenize
#lets try the same with tokenization
# lets first tokenize  a simple sentence
sent = 'Hello Friends ! , How are you? I hope, you like Python Programming.'
#this will create the list of sentences in the text.
#It is also called as segmentation.
sent_tokenize(sent)
```

```python
#now we can also break the segmented sentences into tokes
word_tokenize(sent)

```

```python
#if we solve the problem using tokenization approach, even if we add a comma operator after the integer
#the tokenizer will seperate the comma and the digits hence we can find the digits to find the average
text = "They told that their ages are 21, 25, and 23 respectively"
tokens = word_tokenize(text)
print(tokens)
```

```python
ages=[]
for word in tokens:
  if word.isdigit():
      ages.append(int(word))
average_age = sum(ages) / len(ages)
print(average_age)
```

# ***Lemmentization***

```python
from nltk.stem import WordNetLemmatizer
```

```python
wnl = WordNetLemmatizer()
```

```python
word = 'players'
print('Noun:',wnl.lemmatize(word,pos = 'n'))
print('Verb:',wnl.lemmatize(word,pos = 'v'))
print('Adjective:',wnl.lemmatize(word,pos = 'a'))
print('Adverb:',wnl.lemmatize(word,pos = 'r'))
```

```python
word = 'worst'
print('Noun:',wnl.lemmatize(word,pos = 'n'))
print('Verb:',wnl.lemmatize(word,pos = 'v'))
print('Adjective:',wnl.lemmatize(word,pos = 'a'))
print('Adverb:',wnl.lemmatize(word,pos = 'r'))
```

####pos tagging

```python
sent ='hello friend !how are you ? I hope, you like python programming '
from nltk import pos_tag
```

```python
tags= pos_tag(word_tokenize(sent))
```

```python
tags
```

```python
for words in tags:
  if words[1].startswith('N'):
    print(words)
```

```python
#to get nouns from the extracted tags
for word,tag in tags:
  if tag.startswith('N'):
    print(word)
```
