```python
url=urllib.request.urlopen('https://en.wikipedia.org/wiki/Larry_O%27Brien')
html=url.read()
```

```python
print(html)
```

```python
from bs4 import BeautifulSoup
# we use beaufifulsoup to remove all the tags from the html page s
#by this we get only the data we need to process
```

```python
soup=BeautifulSoup(html,'html')
```

```python
text=soup.get_text(strip=False)
```

```python
print(text)
```

```python
#tokensization
#puncutations
#stop Words
#stemming
#find top 3 words
```

```python
from nltk.tokenize import word_tokenize
```

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
token=word_tokenize(text)
```

```python
from nltk import FreqDist
```

```python
freq=FreqDist(token)
```

```python
freq.plot(20)
```

```python
from nltk.corpus import stopwords
```

```python
swords = stopwords.words('english')
```

```python
#punctuation removal
token2 = [x for x in token if x.isalpha()]#to see if the token is an alphabet
```

```python
token3 = [x.lower() for x in token2 if x.lower() not in swords]
freq=FreqDist(token3)
```

```python
freq.plot(20,marker = 'o',mfc = 'red')
```

```python
freq.most_common(20)
```

```python
# stemming the words from the tokesn
from  nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
token4=[ps.stem(x) for x in token]
```
