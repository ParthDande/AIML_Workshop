Use of Non-Linear Regressin like Decision Tree Regression is used

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df =pd.read_csv('/content/petrol_consumption.csv')
```

```python
df
```

```python
df.corrwith(df['Petrol_Consumption'])
```

```python
x = df.drop('Petrol_Consumption',axis = 1)
y = df['Petrol_Consumption']
x.shape
```

```python
x.describe()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state =0)
x_train.shape
```

```python
x_test.shape
```

```python
#building the model
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor = DecisionTreeRegressor()
```

```python
regressor.fit(x_train,y_train)
```

```python
from sklearn.tree import plot_tree
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize = (20,14))
plot_tree(regressor,fontsize=8,filled=True,feature_names=x.columns,rounded=True)
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error
```

```python
y_pred = regressor.predict(x_test)
```

```python
result = pd.DataFrame({
    'Actual':y_test,
    'Predicted':y_pred
})
```

```python
result
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test,y_pred)#thireen percent values are matching
```

```python
np.mean(abs(y_test-y_pred))
```

```python
#root mean squared error
np.sqrt(np.mean((y_test-y_pred)**2))
```

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_test,y_pred))
```
