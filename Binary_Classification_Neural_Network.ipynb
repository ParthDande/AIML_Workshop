In this code we will use binary classification
For binary classification our output node is a single node and we use sigoid activation function.
. When we predict the output we will get and arry of probablities , so if the output proabablity is above 0.5 it will be class 1 and below 0.5 will be class 0

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from keras.models import Sequential
from tensorflow import keras
from keras.layers import Dense
```

```python
df = pd.read_csv('/content/banknotes.csv')
```

```python
x= df.drop('Class',axis = 1)
```

```python
y = df['Class']
```

```python
#model = keras.Sequential([layers.Dense(64,activation = 'relu',input_shape=(784,)),layers.Dense(10)])
model = Sequential()
```

```python
#adding layers to the model
layer1= Dense(4,input_shape=(4,),activation = 'relu')
model.add(layer1)
```

```python
layer2 = Dense(16,activation = 'relu')
model.add(layer2)
```

```python
layer3 = Dense(8,activation = 'relu')
model.add(layer3)
```

```python
layer4 = Dense(1,activation = 'sigmoid')
model.add(layer4)
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x,y,epochs = 10,batch_size = 5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 0)
x.shape
```

```python
model.fit(x_train,y_train,epochs = 10,batch_size = 6)
```

```python
model.evaluate(x_test,y_test,batch_size =6,verbose=0)
```
