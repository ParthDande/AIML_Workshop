**Data visualization using Matplotlib**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.array([3,8,12,14,15,24,38,42,49,55])
y = np.array([11,16,34,47,61,67,69,75,88,91,])
z=np.array([76,74,66,54,43,32,26,28,21,13])
p= np.array([56,66,44,34,67,66,52,74,37,66])
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,z)
```

```python
plt.plot(x,p,color = 'red')
```

```python
plt.title('My first Plot')
plt.grid()
plt.xlim(0,70)
plt.ylim(0,100)
plt.yticks(range(0,100,10))
plt.plot(x,y)
```

```python
plt.title('My first Plot')
plt.grid()
plt.xlim(0,70)
plt.ylim(0,100)
plt.yticks(range(0,100,10))
plt.plot(x,y,color = 'green',marker = 'o',mfc = 'red',lw=10)
```

```python
df = pd.read_csv('/content/student.csv')
plt.title('Students vs Marks')
plt.xlabel('Roll No')
plt.ylabel('Marks')
plt.grid()
plt.xlim(0,11)
plt.ylim(50,100)
plt.plot(df['roll'],df['marks'],color='blue',marker ='o')
```

```python
plt.title('Students vs Marks')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.xlim(0,70)
plt.ylim(0,100)
plt.yticks(range(0,100,10))
plt.plot(x,y,color = 'green',label = 2023)
plt.plot(x,z,color = 'red',label = 2024)
plt.legend()
```

```python
plt.subplot(1,2,1)#one row 2 columns and the first plot
plt.plot(x,y)
plt.subplot(1,2,2)#one row 2 columns second plot
plt.plot(x,z,color = 'red')
plt.show()
```

Scatter plot

```python
plt.figure(figsize=(16,9))

plt.scatter(x,y,color = 'red',marker = 'x')
```

Bar Plot

```python
plt.figure(figsize=(16,9))
plt.bar(x,y,color='maroon')
```
