softmax activation function is used as probablity distrubtion.

It is used for multiclass format.
It assigns probablity to the outputs and the high probablity class is predicted

For single class classification problem we use sigmoid activation fucntion.

Relu activaion funtion only forwards positive value .

It is mainly used in hidden layers of a neural Network
for regression probelems we dont have a activation function and we need only 1 neuron in the output layer.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)
def softmax(z):
  val = np.exp(z)/np.sum(np.exp(z))
  return val
```

```python
plt.figure(figsize = (16,9))
plt.subplot(2,2,1)
plt.plot(x,relu(x),label = 'relu')
plt.subplot(2,2,2)
plt.plot(x,tanh(x),color ='red',label='tanh')#tan h is same as sigmoid but its range is -1 to 1
plt.subplot(2,2,3)
plt.plot(x,sigmoid(x),color = 'green',label = 'sigmoid')
plt.legend()
```
